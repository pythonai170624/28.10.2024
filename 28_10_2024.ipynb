```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]  # X-axis values
y = [2, 4, 6, 8, 10]  # Y-axis values

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = 2x')

# Add title and labels
plt.title("Line Graph Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show legend
plt.legend()

# Display the plot
plt.show()
```
